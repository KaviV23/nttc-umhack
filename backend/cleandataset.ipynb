```python
import polars as pl
from pathlib import Path
from skimpy import skim

# Try climbing up one or two directories if needed
file_path = Path("../backend/data/derivative_datasets/combined_orders.csv")  # or ../../ if deeper
print("Exists:", file_path.exists())

df = pl.read_csv(file_path, try_parse_dates=True)
skim(df)
```

```python
import polars as pl

# First, convert all time strings to proper Datetime format
df = df.with_columns([
    pl.col("order_time").str.strptime(pl.Datetime).alias("order_time_parsed"),
    pl.col("driver_arrival_time").str.strptime(pl.Datetime).alias("driver_arrival_time_parsed"),
    pl.col("driver_pickup_time").str.strptime(pl.Datetime).alias("driver_pickup_time_parsed"),
    pl.col("delivery_time").str.strptime(pl.Datetime).alias("delivery_time_parsed"),
])

# Now extract features from parsed timestamps
df = df.with_columns([
    pl.col("order_time_parsed").dt.date().alias("order_date"),
    pl.col("order_time_parsed").dt.hour().alias("order_hour"),
    pl.col("order_time_parsed").dt.weekday().alias("order_weekday"),
])

# Add duration/delay deltas
df = df.with_columns([
    (pl.col("driver_arrival_time_parsed") - pl.col("order_time_parsed")).alias("time_to_arrive"),
    (pl.col("driver_pickup_time_parsed") - pl.col("driver_arrival_time_parsed")).alias("wait_to_pickup"),
    (pl.col("delivery_time_parsed") - pl.col("driver_pickup_time_parsed")).alias("delivery_duration"),
])

df = df.drop([
    "order_time", "driver_arrival_time", "driver_pickup_time", "delivery_time"
])

df.head(10)
```

```python
import matplotlib.pyplot as plt

# Group by date and count orders
order_volume = (
    df.group_by("order_date")
      .agg(pl.count("order_id").alias("order_count"))
      .sort("order_date")
)

# Convert to pandas for matplotlib
order_volume_pandas = order_volume.to_pandas()

# Plot
plt.figure(figsize=(12, 5))
plt.plot(order_volume_pandas["order_date"], order_volume_pandas["order_count"])
plt.title("Orders per Day")
plt.xlabel("Date")
plt.ylabel("Number of Orders")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import polars as pl

# Load your dataset
df = pl.read_csv("combined_orders.csv", try_parse_dates=True)

# 1. Check for NULLs
null_counts = df.select([
    pl.col(c).is_null().sum().alias(f"{c}_nulls") for c in df.columns
])
print(null_counts)

# 2. Log transform high-skew numeric columns
df = df.with_columns([
    pl.col("order_value").log1p().alias("order_value_log"),
    pl.col("quantity").log1p().alias("quantity_log"),
    pl.col("subtotal").log1p().alias("subtotal_log"),
])

# 3. Outlier clipping (Winsorization-style)
def clip_column(df, col, lower_q=0.01, upper_q=0.99):
    q_lo = df.select(pl.col(col).quantile(lower_q)).item()
    q_hi = df.select(pl.col(col).quantile(upper_q)).item()
    return df.with_columns([
        pl.col(col).clip(lower_bound=q_lo, upper_bound=q_hi).alias(f"{col}_clipped")
    ])

df = clip_column(df, "order_value")
df = clip_column(df, "quantity")
df = clip_column(df, "subtotal")

# 4. Optional: Encode merchant_name if needed for modeling
from sklearn.preprocessing import LabelEncoder

merchant_le = LabelEncoder()
merchant_labels = merchant_le.fit_transform(df["merchant_name"].to_list())
df = df.with_columns(pl.Series("merchant_label", merchant_labels))
```
